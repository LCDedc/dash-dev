## 使用selenium爬取，因为该网页是动态点击的，需要模拟点击事件，做之前如果先检查Chrome浏览器和driver是否版本一致，版本报错参考https://blog.csdn.net/qq_40604853/article/details/81388078 还有外网比较慢，可以不用先爬五年的，页数太多

```python
from selenium import webdriver
import time
import json
import pandas as pd
import tqdm
```

```python
url="https://www.nasdaq.com/market-activity/stocks/amzn/historical"
driver = webdriver.Chrome()
#3、driver发生get请求
driver.get(url)
for i in tqdm.tqdm_notebook(range(71)):
    if i==0:        
        #第一页
        #进入五年模式,找到5年这个按钮对应的源代码，将xpath复制下来（可以问我）
        element=driver.find_element_by_xpath('/html/body/div[2]/div/main/div[2]/div[4]/div[3]/div/div[1]/div/div[1]/div[3]/div/div/div/button[5]')
        driver.execute_script("arguments[0].click();", element)
        #外网加载速度比较慢，手动阻塞，以免响应不及时
        time.sleep(10)
        #保存网页源代码
        data = driver.page_source
        a=pd.read_html(data)#提取表格
        df_stocks=pd.DataFrame(a[0])#变为数据框
    else:
        #第二页以后需要点击下一页，同样将下一页的xpath复制下来
        element=driver.find_element_by_xpath('/html/body/div[2]/div/main/div[2]/div[4]/div[3]/div/div[1]/div/div[1]/div[5]/button[2]')
        driver.execute_script("arguments[0].click();", element)
        #外网加载速度比较慢，手动阻塞，以免响应不及时
        time.sleep(5)
        #保存网页源代码
        data = driver.page_source
        a=pd.read_html(data)
        df_stocks=pd.concat([df_stocks,pd.DataFrame(a[0])])#将多页的数据拼接到一个数据框里面
```
